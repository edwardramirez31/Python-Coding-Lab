```python
import math

def sen(grads):
    return math.sin(math.radians(grads))
def cos(grads):
    return math.cos(math.radians(grads))
def tan(grads):
    return math.tan(math.radians(grads))
import numpy as np
import matplotlib.pyplot as plt
```

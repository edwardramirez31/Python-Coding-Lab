```python
import numpy as np
```

```python
array = np.array([1, 2, 4])
array
```

```python
array = np.array([
    [1, 4, 3],
    [5, 4, 6],
    [8, 2, 1]
])
array
```

```python
# transpuesta
array.T
```

```python
inversa = np.linalg.inv(array)
inversa
```

```python
determinante = np.linalg.det(array)
determinante
```

```python
# Ax = b
b = np.array([1, 2, 4])
b
```

```python
#x = A-1 * b
x = inversa * b
x
```

```python
x = inversa.dot(b)
x
```

```python
# diagonal
diag = np.diag(inversa)
diag
```

```python
x = np.linspace(0, 401, 1000)
x
```

```python
np.arange(5, 100, 2)
```

```python
# desde 0 a 150 y = x, desde 150 a 300 y = x **2, desde 300 a 401, y = x ** 2 + 2 * x + 1
filtro = x >= 200
filtro
```

```python
x[filtro]
```

```python
y = x[(x >= 0) & (x < 150)]
y
```

```python
y = x[(x >= 150) & (x < 300)] ** 2
y
```

```python
# desde 0 a 150 y = x, desde 150 a 300 y = x *999*2, desde 300 a 401, y = x ** 2 + 2 * x + 1
y = np.piecewise(x, [((x >= 0) & (x < 150)), ((x >= 150) & (x < 300)), ((x >= 300))], [1, 2, -1])
y
```

```python
# desde 0 a 150 y = x, desde 150 a 300 y = x *999*2, desde 300 a 401, y = x ** 2 + 2 * x + 1
y = np.piecewise(x, [((x >= 0) & (x < 150)), ((x >= 150) & (x < 300)), ((x >= 300))], [lambda x: x, lambda x: -x ** 2, lambda x: x ** 2 + 2 * x + 1])
y
```

```python
import numpy as np
np.set_printoptions(precision=3,suppress=True)
```

```python
x = np.arange(0,501)
y = np.array([])
x
```

```python
Contador = 0
while Contador <= 500:
    if Contador <= 100:
        y1 = (1/3)*x[Contador]**2-20
        y = np.append(y,y1)
    
    elif 100 < Contador <=350:
        y2 = -1/3*x[Contador]+5
        y = np.append(y,y2)
   
    else:
        y3 =-5*x[Contador]+2 
        y = np.append(y,y3)
    
    Contador = Contador + 1

y
```

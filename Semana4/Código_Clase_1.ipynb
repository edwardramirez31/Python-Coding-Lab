## ¿CÓMO AGREGO LEYENDA?
1. Colocar un argumento `label` dentro de la función `plot`
2. `plt.legend` 

```python
x = np.linspace(-10, 10, 1001)
y = x ** 2

datos_x_2 = np.linspace(-10, 10, 1001)
datos_y_2 = 2 * datos_x_2 + 10
```

```python
plt.plot(x, y, label='x ** 2', color='#d68354', linewidth=2, linestyle='--')
plt.plot(datos_x_2, datos_y_2, label='2 * x + 10', color='#664dd6', linewidth=2, linestyle='-.')
plt.xlabel('x')
plt.ylabel('F(x)')
plt.title('Gráfica de x al cuadrado')
plt.grid()
plt.legend()
plt.tight_layout()
plt.show()
```

# EJEMPLO PRÁCTICO
El objetivo es calcular con facilidad los valores de los esfuerzos cortantes V(x) y los momentos flectores M(x) en cualquier punto de la viga, y su representación en función de la distancia x  a uno de los extremos. Como ejemplo analizamos una viga AB de longitud L (figura a), simplemente apoyada y bajo la acción de una sola carga P concentrada en el punto medio.

<img src='viga.png'>
<p style="text-align: center">Cortante tramo 1</p>

$$ V(x) = \frac{P}{2} $$
<p style="text-align: center">Cortante tramo 2</p>

$$ V(x) = \frac{-P}{2} $$
<p style="text-align: center">Momento tramo 1</p>

$$ M_1(x) = \frac{P}{2} * x  $$
<p style="text-align: center">Cortante tramo 2</p>

$$ M_2(x) = \frac{P}{2} * (L - x)  $$

## SOLUCIÓN
1. Determinar u obtener los datos en el eje x
2. Obtener su equivalente en el eje y. Reemplazar x en la funcion
3. Utilizar `plot`
4. Crear graficos de relleno `fill_between`
5. Personalizar mis plots

```python
# en metros
LONGITUD = 4
# en kN
CARGA_P = 100
```

```python
eje_x = np.linspace(0, LONGITUD, 1001)
```

```python
# 1. CORTANTE
cortante = np.full(1001, CARGA_P / 2)

plt.plot(eje_x[eje_x <=2], cortante[eje_x <= 2])
plt.fill_between(eje_x[eje_x <= 2], cortante[eje_x <= 2], alpha=0.25)

plt.plot(eje_x[eje_x >= 2], -cortante[eje_x >=2])
plt.fill_between(eje_x[eje_x >= 2], -cortante[eje_x >= 2], alpha=0.25)

plt.xlabel('x [m]')
plt.ylabel('V(x) [kN]')
plt.yticks(np.arange(-50, 60, 10))
plt.xticks(np.arange(0, 5))
plt.title('Diagrama de fuerza cortante [kN]')
```

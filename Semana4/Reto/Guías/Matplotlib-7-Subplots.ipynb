```python
%matplotlib inline
import pandas as pd
import matplotlib.pyplot as plt
```

```python
dataframe = pd.read_csv('salaries.csv')
dataframe
```

```python
ages = dataframe['Age']
all_devs = dataframe['All_Devs']
python = dataframe['Python']
javascript = dataframe['JavaScript']
```

```python
plt.style.use('seaborn')

plt.plot(ages, python, label='Python')
plt.plot(ages, javascript, color='r', label='JavaScript')
plt.plot(ages, all_devs, linestyle='--', color='#444444', linewidth=2, label='All Devs')


plt.legend()

plt.title('Developer Salary (USD) By Age')
plt.xlabel('Age')
plt.ylabel('Salary (USD)')

plt.tight_layout()
plt.show
```

```python
plt.style.use('seaborn')

fig, ax = plt.subplots()

ax.plot(ages, python, label='Python')
ax.plot(ages, javascript, color='r', label='JavaScript')
ax.plot(ages, all_devs, linestyle='--', color='#444444', linewidth=2, label='All Devs')


ax.legend()

ax.set_title('Developer Salary (USD) By Age')
ax.set_xlabel('Age')
ax.set_ylabel('Salary (USD)')

plt.tight_layout()
plt.show
```

```python
fig, ax = plt.subplots(nrows=2, ncols=1)
ax
```

```python
fig, ax = plt.subplots(nrows=4, ncols=3)
ax
```

```python
fig, (ax1, ax2) = plt.subplots(nrows=2, ncols=1, sharex=True)

ax1.plot(ages, all_devs, color='#444444', linestyle='--', linewidth=2, label='All Devs')

ax2.plot(ages, python, label='Python Devs')
ax2.plot(ages, javascript, label='JS Devs')

ax1.legend()
ax2.legend()

ax1.set_title('Developer Salary (USD) By Age')
ax1.set_ylabel('Salary (USD)')

ax2.set_xlabel('Ages')
ax2.set_ylabel('Salary (USD)')

plt.tight_layout()
plt.show
```

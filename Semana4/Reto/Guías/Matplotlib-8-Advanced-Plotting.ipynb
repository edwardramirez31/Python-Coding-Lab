```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(-np.pi, np.pi, 256)
x
```

```python
cos, sen = np.cos(x), np.sin(x)
```

```python
plt.plot(x, cos)
plt.plot(x, sen)

plt.show
```

```python
#plt.style.use('ggplot')

plt.plot(x, cos, linewidth=2, linestyle='-', label='cos(x)')
plt.plot(x, sen, linewidth=2, linestyle='--', label='sen(x)')

#lims
plt.xlim(x.min() * 1.1, x.max() * 1.1)
plt.ylim(cos.min() * 1.1, cos.max() * 1.1)

# ticks
plt.xticks(np.arange(-4, 5))
plt.yticks(np.linspace(-1, 1, 5))
#legend
plt.legend()
# Text
plt.title("Sen(X) y Cos(X)")
plt.xlabel('x')
plt.ylabel('y')

plt.tight_layout()
plt.show
```

```python
#plt.style.use('ggplot')

plt.plot(x, cos, linewidth=2, linestyle='-', label='cos(x)')
plt.plot(x, sen, linewidth=2, linestyle='--', label='sen(x)')

#lims
plt.xlim(x.min() * 1.1, x.max() * 1.1)
plt.ylim(cos.min() * 1.1, cos.max() * 1.1)

# ticks
plt.xticks(ticks=[-np.pi, -np.pi/2, 0, np.pi/2, np.pi], labels=[r'$-\pi$', r'$-\pi/2$', r'$0$', r'$+\pi/2$', r'$+\pi$'])
plt.yticks(ticks=np.linspace(-1, 1, 3), labels=[r'$-1$', r'$0$', r'$+1$'])
#legend
plt.legend()
# Text
plt.title("Sen(X) y Cos(X)")
plt.xlabel('x')
plt.ylabel('y')

plt.tight_layout()
plt.show
```

Mover los ejes moviendo los spines, los cuales son las lineas que contienen las etiquetas de los ejes.
1. Obtener el eje
2. Acceder al objeto que contiene la informacin de ese spine mediante el diccionario spines. La llave es la posicin
3. Para establecer el color, se usa el mtodo set_color
4. Para establecer la posicin, se usa el mtodo set_position y se le pasa una tupla donde le primero parmetro es 'data' y el segundo el lugar de la grfica donde quiero colocar el spin
5. Acceder al eje x y y mediante `xaxis` y `yaxis`
6. Usar el mtodo `set_ticks_position` y pasar el lugar donde quiero que se coloquen las etiquetas

```python
axis = plt.gca()

# get spines
axis.spines['top'].set_color('none')
axis.spines['right'].set_color('none')
axis.spines['bottom'].set_color('#000000')
axis.spines['left'].set_color('#000000')
axis.spines['bottom'].set_position(('data', 0))
axis.spines['left'].set_position(('data', 0))
#axis ticks postion
axis.xaxis.set_ticks_position('bottom')
axis.yaxis.set_ticks_position('left')

#plt.style.use('seaborn-notebook')

plt.plot(x, cos, linewidth=2, linestyle='-', label='cos(x)')
plt.plot(x, sen, linewidth=2, linestyle='--', label='sen(x)')

#lims
plt.xlim(x.min() * 1.1, x.max() * 1.1)
plt.ylim(cos.min() * 1.1, cos.max() * 1.1)

# ticks
plt.xticks(ticks=[-np.pi, -np.pi/2, 0, np.pi/2, np.pi], labels=[r'$-\pi$', r'$-\pi/2$', r'$0$', r'$+\pi/2$', r'$+\pi$'])
plt.yticks(ticks=np.linspace(-1, 1, 3), labels=[r'$-1$', r'$0$', r'$+1$'])
#legend
plt.legend(loc='upper left')
# Text
plt.title("Sen(X) y Cos(X)")
plt.xlabel('x')
plt.ylabel('y')

plt.tight_layout()
plt.show
```

Dibujar un marcador y tambin una lnea de puntos
1. La lnea de puntos se hace con un plot

```python
axis = plt.gca()

# get spines
axis.spines['top'].set_color('none')
axis.spines['right'].set_color('none')
axis.spines['bottom'].set_color('#000000')
axis.spines['left'].set_color('#000000')
axis.spines['bottom'].set_position(('data', 0))
axis.spines['left'].set_position(('data', 0))
#axis ticks postion
axis.xaxis.set_ticks_position('bottom')
axis.yaxis.set_ticks_position('left')

#plt.style.use('seaborn-notebook')

plt.plot(x, cos, linewidth=2, linestyle='-', label='cos(x)')
plt.plot(x, sen, linewidth=2, linestyle='-', color='red', label='sen(x)')

#lims
plt.xlim(x.min() * 1.1, x.max() * 1.1)
plt.ylim(cos.min() * 1.1, cos.max() * 1.1)

# ticks
plt.xticks(ticks=[-np.pi, -np.pi/2, 0, np.pi/2, np.pi], labels=[r'$-\pi$', r'$-\pi/2$', r'$0$', r'$+\pi/2$', r'$+\pi$'])
plt.yticks(ticks=np.linspace(-1, 1, 3), labels=[r'$-1$', r'$0$', r'$+1$'])
#legend
plt.legend(loc='upper left')
# Text
plt.title("Sen(X) y Cos(X)")
plt.xlabel('x')
plt.ylabel('y')

x_value = np.pi / 2
# line
plt.plot([x_value, x_value], [0, np.sin(x_value)],color='red', linestyle='--')
# point
plt.scatter([x_value, ],[np.sin(x_value), ], 50, color='red')
# anotacin
plt.annotate(r'$sin(\frac{\pi}{2})=1$',
            xy=(x_value, np.sin(x_value)), xycoords='data',
            xytext=(+10, +30), textcoords='offset points', fontsize=16,
            arrowprops=dict(arrowstyle="->", connectionstyle="arc3,rad=.2"))

plt.tight_layout()
plt.show
```

- arguments: https://matplotlib.org/1.5.3/users/annotations_guide.html
- annotate: https://matplotlib.org/2.0.2/api/pyplot_api.html

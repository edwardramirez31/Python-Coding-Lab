# FILTRAR FILAS
A menudo se rquieren filtrar las filas que cumplen cierto criterio o condición en una columna en especifico. Para ello, se crean `filtros` los cuales se componen de un objeto tipo Series, valores deseados y operadores condicionales.

> Si quiero obtener un dataframe con base en uno o más valores de una columna especifica, uso este tipo de filtros. Estos filtros devuelven Series donde cada fila tiene su resultado correspondiente True o False

```python
df['last'] == 'Doe'
```

Dado que pasando Series de booleanos a un dataframe puedo filtrarlo, existen dos maneras para asignar el `filtro` al DataFrame y obtener un subconjunto de este:
1. Pasar el filtro como si se estuviese tratando de obtener una columna: df[filterName]
2. Pasar el filtro dentro del primer parámetro de loc: df.loc[filter]

> Cuando asigno el filtro al DataFrame, obtengo un sub DataFrame, filtrado por las filas que cumplen con el criterio deseado

```python
filt = df['first'] == 'Edward'
df[filt]
```

```python
df.loc[filt]
```

```python
# Resulta mejor usar loc dado que puedo filtrar tambien por columnas, en caso de que lo requiera
df.loc[filt, 'email']
```

## AND Y OR:
También se pueden filtrar filas con filtros más complejos de la misma forma que se hace con los condicionales en Python, a excepción que:
* Se debe usar el símbolo `&` en lugar de `and`
* Se debe usar el símbolo `|` en lugar de `or`
* Se debe usar el símbolo `-` en lugar de `not`

```python
# quiero todas las filas que contengan 'Edward' como primero nombre o "johndoe@gmail.com" como email
orFilter = (df['first'] == 'Edward') | (df['email'] == 'johndoe@gmail.com')
df.loc[orFilter]
```

```python
andFilter = (df['first'] == 'John') & (df['email'] == 'johndoe@gmail.com')
df.loc[andFilter]
```

```python
# Si quiero obtener todo lo que no cumplela condición 'andFilter' uso -
# Todos las filas que no tienen el nombre John y el email johndoe@gmail.com
df.loc[-andFilter]
```

```python
# Quiero filtrar nuevamente bajo esta condición de salario, pero que aparezcan solo los países USA, United Kingdom, Germany,
# Canada and India
countries = ['United States', 'United Kingdom', 'Canada', 'India', 'Germany']
# Al ser varios países, puedo crear una lista y pasarla en el isin method
filt = survey_df['Country'].isin(countries) & filt
survey_df.loc[filt, ['Country', 'LanguageWorkedWith', 'ConvertedComp']]
```

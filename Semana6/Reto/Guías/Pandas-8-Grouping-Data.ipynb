# Grouping
Si quiero ver ciertos resultados con base en cierta columna, se usa la función `groupby`. A esta función se pasa una lista de columnas con base en las cuales se quiere agrupar
- Split object
- Appling a function
- Combining Results

```python
# I want the most popular social media by country
country_group = df.groupby(['Country'])
country_group
```

Al final voy a obtener un objeto `DataFrameGrupoBy` el cual contiene diferentes grupos. *Con el método `groupby` se separó el dataframe en diferentes grupos (similar a split)*
> Para acceder a cada grupo uso el método `get_group` pasando uno de los valores de la columna con base en la cual se dividió el dataframe

```python
# Grab a specific group by contry name
country_group.get_group('Colombia')
```

```python
filt = df['Country'] == 'India'
df.loc[filt]['LanguageWorkedWith'].str.contains('Python')
```

```python
# Count the number of people with the sum function
# It counts all the trues as 1
df.loc[filt]['LanguageWorkedWith'].str.contains('Python').sum()
```

```python
country_group['LanguageWorkedWith'].str.contains('Python').sum()
```

```python
# use apply to run a function in each series of that group
python_total = country_group['LanguageWorkedWith'].apply(lambda x: x.str.contains('Python').sum())
```

```python
country_group.get_group('India')['LanguageWorkedWith'].str.contains('Python').sum()
```

```python
# What percentage of people from each country know Python?
total = df['Country'].value_counts()
```

```python
python_total / total * 100
```

```python
# combine series with concat method
# by default concats on rows
python_df = pd.concat([total, python_total], axis='columns', sort=False)
python_df
```

```python
python_df.rename(columns={'Country': 'TotalPeople', 'LanguageWorkedWith': 'TotalUsePython'}, inplace=True)
```

```python
python_df
```

```python
python_df['Percentage'] = python_df['TotalUsePython'] / python_df['TotalPeople'] * 100
```
